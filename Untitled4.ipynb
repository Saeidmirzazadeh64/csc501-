```python
from tensorflow import keras
from tensorflow.keras import layers

num_actions = 18

def create_q_model():
    # Network defined by the Deepmind paper
    inputs = layers.Input(shape=(84, 84, 1,))
    
    layer1 = layers.Conv2D(32, 8, strides=4, activation="relu")(inputs)
    layer2 = layers.Conv2D(64, 4, strides=2, activation="relu")(layer1)
    layer3 = layers.Conv2D(64, 3, strides=1, activation="relu")(layer2)
    layer4 = layers.Flatten()(layer3)
    layer5 = layers.Dense(512, activation="relu")(layer4)
    action = layers.Dense(num_actions, activation="linear")(layer5)

    return keras.Model(inputs=inputs, outputs=action)

# The first model makes the predictions for Q-values which are used to make an action.
model = create_q_model()
# Target model
model_target = create_q_model()
model.summary()
```

```python
gamma = 0.99  # Discount factor for past rewards

# Setting epsilon decay parameters
epsilon = 1.0  
epsilon_max_1 = 1.0 
epsilon_min_1 = 0.2  
epsilon_max_2 = epsilon_min_1  
epsilon_min_2 = 0.1
epsilon_max_3 = epsilon_min_2  
epsilon_min_3 = 0.02

epsilon_interval_1 = (epsilon_max_1 - epsilon_min_1)  
epsilon_interval_2 = (epsilon_max_2 - epsilon_min_2)  
epsilon_interval_3 = (epsilon_max_3 - epsilon_min_3)  

# Number of frames for exploration
epsilon_greedy_frames = 1000000.0

# Number of frames to take random action and observe output
epsilon_random_frames = 50000

# Maximum Replay Buffer volume
max_memory_length = 190000

# Size of batch taken from replay buffer
batch_size = 32  
max_steps_per_episode = 10000

# Train the model after 20 actions
update_after_actions = 20

# How often to update the target network
update_target_network = 10000

# In the Deepmind paper they use RMSProp however then Adam optimizer improves training time
optimizer = keras.optimizers.Adam(learning_rate=0.00025, clipnorm=1.0)

# Using huber loss for stability
loss_function = keras.losses.Huber()
```

```python
import cv2
import numpy as np
import tensorflow as tf

def preprocess_state(state):
    # Assuming 'state' is an image represented as a NumPy array
    # Resize the image to a new height and width
    new_height, new_width = 84, 84  # Adjust these dimensions as needed
    state = cv2.resize(state, (new_width, new_height))
    
    # Check if the image is not already in grayscale
    if len(state.shape) > 2:
        # Convert the image to grayscale
        state = cv2.cvtColor(state, cv2.COLOR_RGB2GRAY)
    
    # Normalize pixel values to the range [0, 1]
    state = state / 255.0
    
    # Add any additional preprocessing steps if necessary
    
    return state

# Experience replay buffers
action_history = []
state_history = []
state_next_history = []
rewards_history = []
done_history = []

episode_reward_history = []
running_reward = 0
episode_count = 0
frame_count = 0


learning_rate = 0.001  # Adjust this value as needed

optimizer = tf.keras.optimizers.Adam(learning_rate)

# ... (previous code)

max_episodes = 20  # Increase the number of episodes
max_steps_per_episode = 100  # Increase the number of steps per episode



while episode_count < max_episodes:
    print("episode:", episode_count)
    state_tuple = env.reset()
    state = np.array(state_tuple[0])
    state = preprocess_state(state)
    print("Initial state shape:", state.shape)
    episode_reward = 0

    for timestep in range(1, max_steps_per_episode):
        frame_count += 1

        # ... (previous code)

        # Use epsilon-greedy for exploration
        if frame_count < epsilon_random_frames or epsilon > np.random.rand(1)[0]:
            # Take random action
            action = np.random.choice(num_actions)
        else:
            # Predict action Q-values
            # From environment state
            state_tensor = tf.convert_to_tensor(state, dtype=tf.float32)
            state_tensor = tf.expand_dims(state_tensor, 0)
            state_tensor = tf.expand_dims(state_tensor, -1)  # Add this line to match the input shape (84, 84, 1)
            action_probs = model(state_tensor, training=False)
            # Take the best action
            action = tf.argmax(action_probs[0]).numpy()

        # ... (previous code)

        # Apply the sampled action in our environment
        state_next, reward, done, _, _ = env.step(action)
        state_next = np.array(state_next[0])  # Assuming the relevant data is at the first position
        state_next = preprocess_state(state_next)  # Add this line to preprocess the state

        episode_reward += reward

        # Save actions and states in the replay buffer
        action_history.append(action)
        state_history.append(state)
        state_next_history.append(state_next)
        done_history.append(done)
        rewards_history.append(reward)
        state = state_next

        # Update every 20th frame and once batch size is over 32
        if frame_count % update_after_actions == 0 and len(done_history) > batch_size:
            # Get indices of samples for replay buffers
            indices = np.random.choice(range(len(done_history)), size=batch_size)

            # Using list comprehension to sample from the replay buffer
            state_sample = np.array([state_history[i] for i in indices])
            state_next_sample = np.array([state_next_history[i] for i in indices])
            rewards_sample = [rewards_history[i] for i in indices]
            action_sample = [action_history[i] for i in indices]
            done_sample = tf.convert_to_tensor(
                [float(done_history[i]) for i in indices]
            )

            # Build the updated Q-values for the sampled future states
            # Use the target model for stability
            future_rewards = model_target.predict(state_next_sample)
            # Q value = reward + discount factor * expected future reward
            updated_q_values = rewards_sample + gamma * tf.reduce_max(
                future_rewards, axis=1
            )

            # If the final frame, set the last value to -1
            updated_q_values = updated_q_values * (1 - done_sample) - done_sample

            # Create a mask so we only calculate loss on the updated Q-values
            masks = tf.one_hot(action_sample, num_actions)

            with tf.GradientTape(persistent=True) as tape:
                # Train the model on the states and updated Q-values
                q_values = model(state_sample)

                # Apply the masks to the Q-values to get the Q-value for the action taken
                q_action = tf.reduce_sum(tf.multiply(q_values, masks), axis=1)
                # Calculate loss between new Q-value and old Q-value
                loss = loss_function(updated_q_values, q_action)

            # Backpropagation
            grads = tape.gradient(loss, model.trainable_variables)
            optimizer.apply_gradients(zip(grads, model.trainable_variables))

        if frame_count % update_target_network == 0:
            # update the target network with new weights
            model_target.set_weights(model.get_weights())
            # Log details
            template = "running reward: {:.2f} at episode {}, frame count {}, epsilon {:.3f}, loss {:.5f}"
            print(template.format(running_reward, episode_count, frame_count, epsilon, loss))

        # Limit the state and reward history
        if len(rewards_history) > max_memory_length:
            del rewards_history[:1]
            del state_history[:1]
            del state_next_history[:1]
            del action_history[:1]
            del done_history[:1]

        if done:
            break

    # Update running reward to check the condition for solving
    episode_reward_history.append(episode_reward)
    if len(episode_reward_history) > 100:
        del episode_reward_history[:1]
    running_reward = np.mean(episode_reward_history)

    print("Episode {}: Total Reward: {}".format(episode_count, episode_reward))

    episode_count += 1

    if running_reward > 18:  # Condition to consider the task solved
        print("Solved at episode {}!".format(episode_count))
        break
```

```python
# Function to test the trained agent
def test_agent(env, model, num_episodes=5):
    for episode in range(num_episodes):
        # Setup
        done = False
        score = 0

        # Reset environment and get initial observation
        obs_tuple = env.reset()
        obs = preprocess_state(np.array(obs_tuple[0]))

        # RL Loop
        while not done:
            # Choose action using the trained model (exploitation)
            state_tensor = tf.convert_to_tensor(obs, dtype=tf.float32)
            state_tensor = tf.expand_dims(state_tensor, 0)
            state_tensor = tf.expand_dims(state_tensor, -1)
            action_probs = model(state_tensor, training=False)
            action = tf.argmax(action_probs[0]).numpy()

            # Take action
            obs_tuple, reward, done, _, _ = env.step(action)
            obs = preprocess_state(np.array(obs_tuple[0]))

            # Update score
            score += reward

            # Render the environment (you can uncomment this line if rendering is supported)
            # env.render()

        # Print statistics
        print(f"Episode #{episode + 1} - Score: {score}")

# Test the trained agent
test_agent(env, model)
```
